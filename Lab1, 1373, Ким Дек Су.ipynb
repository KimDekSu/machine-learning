```python
#2а. Предметная область: FIFA World Cup (Чемпионат мира по футболу)
#Источник данных: Официальные данные FIFA
#Характер данных: Реальные данные

#2b. Атрибуты и их типы:
#Year - Год проведения (числовой)
#Datetime - Дата и время матча (строковый)
#Stage - Этап турнира (строковый)
#Stadium - Стадион, на котором прошел матч (строковый)
#City - Город, в котором прошел матч (строковый)
#Home.Team.Name - Название домашней команды (строковый)
#Home.Team.Goals - Количество голов, забитых домашней командой (числовой)
#Away.Team.Goals - Количество голов, забитых гостевой командой (числовой)
#Away.Team.Name - Название гостевой команды (строковый)
#Attendance - Количество зрителей на матче (числовой)
#Half.time.Home.Goals - Количество голов домашней команды к перерыву (числовой)
#Half.time.Away.Goals - Количество голов гостевой команды к перерыву (числовой)
#Referee - Имя судьи матча (строковый)
#Assistant.1 - Имя первого ассистента судьи (строковый)
#Assistant.2 - Имя второго ассистента судьи (строковый)
#RoundID - Идентификатор раунда (числовой)
#MatchID - Идентификатор матча (числовой)
#Home.Team.Initials - Инициалы домашней команды (строковый)
#Away.Team.Initials - Инициалы гостевой команды (строковый)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузим данные
data = pd.read_csv('C:/Users/user/Downloads/archive/FIFA_World_Cup_1558_23.csv')
data

# 3a. Средние и стандартные отклонения
attributes = ['Home.Team.Goals', 'Away.Team.Goals', 'Attendance', 'Half.time.Home.Goals', 'Half.time.Away.Goals']
mean_values = data[attributes].mean()
std_dev = data[attributes].std()
print("Средние значения:")
print(mean_values)
print("\nСтандартные отклонения:")
print(std_dev)

# 3b. Гистограммы распределения значений
for attribute in attributes:
    plt.figure(figsize=(8, 6))
    plt.hist(data[attribute], bins=20, edgecolor='k')
    plt.title(f'Histogram of {attribute}')
    plt.xlabel(attribute)
    plt.ylabel('Frequency')
    plt.show()

# 3c. Проверка на пропущенные значения
missing_values = data[attributes].isnull().sum()
print("\nПропущенные значения:")
print(missing_values)

# 4a. Корреляция
correlation_matrix = data[attributes].corr()
print("\nМатрица корреляций: \n", correlation_matrix)

#4b. Корреляцию не имеют: Datetime, Datetime, Stage, Stadium, City, Home.Team.Name, Away.Team.Name, Referee,
#Assistant.1, Assistant.2, Home.Team.Initials, Away.Team.Initials

# 4c. Построение матрицы графиков рассеивания
sns.pairplot(data[attributes])
plt.show()
```
